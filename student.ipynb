```python
#data representing dollar amounts as strings. Clean those


def cln_string(string):
    if not pd.isnull(string):
        return("".join(d for d in str(string) if str(d).isdigit()))
    else:
        return(string)
```

```python
#Its possible to have duplicate movie titles due to the fact that every movie seems to be remade.
#Therefore, check duplicate movie titles by release date
#yrs_movies = tn_movie_budgets_df[['release_date', 'movie']]
#
#for film in yrs_movies.movie:
#    thing = yrs_movies[yrs_movies.movie == film]
#    if thing.release_date.duplicated().any():
#        print(film)
#print('nada')
```

```python
def movie_year_func(df, date_col, movie_col):
    yrs_movies = df[[date_col, movie_col]]
    for film in yrs_movies[movie_col]:
        film_date = yrs_movies[yrs_movies[movie_col] == film]
        if film_date[date_col].duplicated().any():
            return('{} has duplicate releases'.format(film))
    return('No duplicates')
```

##### I'm interested in movies that grossed well based on ROI

ROI is net profit divided by cost of investment:

((net profit)/(cost of investment)) * 100 (to return a percentage)

```python
#This function will only work on the tn_movie_budgets dataframe since its based on that dataframes column names
def roi_func(row):
    return(((row['worldwide_gross'] - row['production_budget'])/row['production_budget'])*100)
```

```python
#Get all gross columns into floats and get rid of any punctuation characters
tn_movie_budgets_df.production_budget = tn_movie_budgets_df.production_budget.apply(cln_string).astype(float)
tn_movie_budgets_df.domestic_gross = tn_movie_budgets_df.domestic_gross.apply(cln_string).astype(float)
```

```python
tn_movie_budgets_df['roi_percent'] = tn_movie_budgets_df.apply(roi_func, axis=1)
```
